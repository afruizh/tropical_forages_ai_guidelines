# Convert Excel Dataset List to Markdown Table

This notebook reads `data/datasets.xlsx` and generates a markdown table with columns **Name** (as a markdown link) and **Description**.

```python
import pandas as pd
from pathlib import Path

# Path to the Excel file
excel_path = Path('../data/datasets.xlsx')

# Read the Excel file
df = pd.read_excel(excel_path)

# Ensure columns 'Name', 'Description', and 'URL' exist
assert all(col in df.columns for col in ['Name', 'Description', 'URL']), 'Missing required columns.'

# Create markdown link for Name
df['Name'] = df.apply(lambda row: f'[{row['Name']}]({row['URL']})', axis=1)

# Select only Name and Description
md_table = '| Name | Description |\n|------|-------------|\n'
for _, row in df.iterrows():
    md_table += f'| {row['Name']} | {row['Description']} |\n'

# Output markdown table
print(md_table)
```
